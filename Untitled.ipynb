```python
for idx in range(7):
    dist_7 = []
    for node in central_node:
        distances = nx.single_source_shortest_path_length(G, node, cutoff=7)
        for key_d in distances:
            if distances[key_d] == idx + 1:
                dist_7.append(key_d)
    dist_7 = set(dist_7)
    
    # 检查是不是到到任意节点的距离都为大于等于idx
    final_dist = []
    for node in dist_7:
        min_dis = 10
        for cen_node in central_node:
            temp_dis = nx.shortest_path_length(G, source=cen_node, target=node)
            if(temp_dis < min_dis):
                min_dis = temp_dis
        if min_dis == idx+1:
            final_dist.append(node)
    print(len(final_dist))
```

```python
from sklearn.decomposition import PCA
import numpy as np
attr = np.zeros([6137, 768])

mean_degree = np.mean([deg for _, deg in G.degree()])
stdev_degree = np.std([deg for _, deg in G.degree()])

idx = 0
for node in G.nodes():
    attr[[idx],[0]] = G.degree(node)
    attr[[idx],[1]] = Gd.in_degree(node)
    attr[[idx],[2]] = Gd.out_degree(node)
    attr[[idx],[3]] = nx.clustering(G, node)
    attr[[idx],[4]] = len(nx.single_source_shortest_path_length(G, node, cutoff=2)) - len(nx.single_source_shortest_path_length(G, node, cutoff=1))
    attr[[idx],[5]] = len(nx.single_source_shortest_path_length(G, node, cutoff=3)) - len(nx.single_source_shortest_path_length(G, node, cutoff=2))
    neighbor_degrees = [G.degree(d) for d, _ in G.adj[node].items()]
    for degree_nei in neighbor_degrees:
        attr[[idx],[5+degree_nei]] += 1
    
    
    idx += 1
    
for idx in range(768):
    mean = np.mean(attr[:,[idx]])
    std = np.std(attr[:,[idx]])
    if std == 0:
        continue
    attr[:,[idx]] = (attr[:,[idx]] - mean) / std
    
pca = PCA(n_components=3)
pca.fit(attr)
```
